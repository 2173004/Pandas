# PLOTTING DARI SUATU PANDAS DATA FRAME ##

## Import Modules ###

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

## Persiapan Data Frame ###

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)
df.head()
```

## Line Plot ##

```python
df.head()
```

```python
df.plot(kind='line') #plotting untuk keseluruhan kolom
```

```python
df[['A', 'B']].plot(kind='line') #plotting untuk kolom tertentu
```

### Bar Plot ###

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar') #untuk menyederhanakan tampilan menjadi 5 baris awal saja
```

```python
df[['A', 'B']].head().plot(kind='barh') #menampilkan bar plot menjadi horizontal
```

### Area Plot ###

```python
df.head()
```

```python
df.plot(kind='area') #memakai data kumulatif
```

```python
df[['A', 'B']].plot(kind='area')
```

### Box Plot ###

```python
df.head()
```

```python
df.plot(kind='box') #berbeda dari model visual yang lain kalau box langsung menampilkan 5 baris awal
```

## Histogram ###

```python
df.head()
```

```python
df.plot(kind='hist') #dipresentasikan oleh warna setiap barisnya
```

```python
df[['A', 'B']].plot(kind='hist') #dokus pada 2 kolom saja
```

## Kernel Density Estimation (KDE) ##

```python
df.head()
```

```python
df.plot(kind='kde') #
```

### Scatter Plot ###

```python
df.head()
```

```python
df.plot(x='A', y='B', kind='scatter') #visualisasi yang sangat cocok untuk korelasi antara 2 variabel atau 2 kolom
```
